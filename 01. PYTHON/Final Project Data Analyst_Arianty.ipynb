#Data Understanding


#People

ID: Customer's unique identifier

Year_Birth: Customer's birth year

Education: Customer's education level

Marital_Status: Customer's marital status

Income: Customer's yearly household income

Kidhome: Number of children in customer's household

Teenhome: Number of teenagers in customer's household

Dt_Customer: Date of customer's enrollment with the company

Recency: Number of days since customer's last purchase

Complain: 1 if the customer complained in the last 2 years, 0 otherwise


##Products

MntWines: Amount spent on wine in last 2 years

MntFruits: Amount spent on fruits in last 2 years

MntMeatProducts: Amount spent on meat in last 2 years

MntFishProducts: Amount spent on fish in last 2 years

MntSweetProducts: Amount spent on sweets in last 2 years

MntGoldProds: Amount spent on gold in last 2 years

#Promotion

NumDealsPurchases: Number of purchases made with a discount

AcceptedCmp1: 1 if customer accepted the offer in the 1st campaign, 0 otherwise

AcceptedCmp2: 1 if customer accepted the offer in the 2nd campaign, 0 otherwise

AcceptedCmp3: 1 if customer accepted the offer in the 3rd campaign, 0 otherwise

AcceptedCmp4: 1 if customer accepted the offer in the 4th campaign, 0 otherwise

AcceptedCmp5: 1 if customer accepted the offer in the 5th campaign, 0 otherwise

Response: 1 if customer accepted the offer in the last campaign, 0 otherwise

#Place

NumWebPurchases: Number of purchases made through the company’s website

NumCatalogPurchases: Number of purchases made using a catalogue

NumStorePurchases: Number of purchases made directly in stores

NumWebVisitsMonth: Number of visits to company’s website in the last month

```python
# Nama-nama kolom untuk kampanye yang diterima
accepted_columns = ['AcceptedCmp1', 'AcceptedCmp2', 'AcceptedCmp3', 'AcceptedCmp4', 'AcceptedCmp5']

# Menghitung jumlah kategori 0 dan 1 untuk setiap kolom
counts_0 = [df_marketing[col].value_counts().get(0, 0) for col in accepted_columns]  # Hitung jumlah 0
counts_1 = [df_marketing[col].value_counts().get(1, 0) for col in accepted_columns]  # Hitung jumlah 1

# Posisi untuk setiap kategori (0 dan 1) untuk setiap kolom
x = range(len(accepted_columns))

# Lebar bar untuk kategori 0 dan 1
width = 0.35

# Membuat histogram dengan kategori 0 di kiri dan kategori 1 di kanan
plt.figure(figsize=(10, 6))

# Menambahkan bar untuk kategori 0 (semua kategori 0 di sisi kiri)
bars_0 = plt.bar(x, counts_0, width, label='Category 0', color='blue')

# Menambahkan bar untuk kategori 1 (semua kategori 1 di sisi kanan, digeser dengan width)
bars_1 = plt.bar([p + width for p in x], counts_1, width, label='Category 1', color='orange')

# Menambahkan angka di atas setiap bar kategori 0
for bar in bars_0:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, height, str(int(height)), ha='center', va='bottom', fontsize=10)

# Menambahkan angka di atas setiap bar kategori 1
for bar in bars_1:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, height, str(int(height)), ha='center', va='bottom', fontsize=10)

# Menambahkan label dan judul
plt.title('Histogram of Accepted Campaigns (Category 0 and 1)')
plt.xlabel('Campaigns')
plt.ylabel('Frequency')

# Menambahkan sumbu x dengan kategori kampanye
plt.xticks([p + width / 2 for p in x], accepted_columns)

# Menambahkan legenda
plt.legend()

# Menampilkan plot
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import scipy.stats as stats
import seaborn as sns
import matplotlib.pyplot as plt
```

##Tahapan Uji Hipotesis:

###Penentuan hipotesis:
H0: Tidak ada korelasi signifikan antara Income dan Monetary

Ha: Ada korelasi signifikan antara Income dan Monetary

###Uji Statistik:
-Uji Pearson Correlation jika data berdistribusi normal.

-Uji Spearman Rank Correlation: jika data tidak berdistribusi normal.


###P-value:

-Jika p-value <=0,05, H0 ditolak.

-Jika p-value >0,05, H0 diterima.


##Cek normalitas

```python
#cek normalitas Income

# Shapiro-Wilk test for normality
stat, p_value = stats.shapiro(df_marketing['Income'].dropna()) # Changed df to df_marketing

print(f"Shapiro-Wilk Test Stat: {stat}")
print(f"P-Value: {p_value}")

# Interpretation
if p_value < 0.05:
    print("The data is not normally distributed (reject H0).")
else:
    print("The data follows a normal distribution (fail to reject H0).")
```

Kesimpulan: Income tidak berdistribusi normal


```python
#cek normalitas Monetary

# Shapiro-Wilk test for normality
stat, p_value = stats.shapiro(df_marketing['Monetary'].dropna()) # Changed df to df_marketing

print(f"Shapiro-Wilk Test Stat: {stat}")
print(f"P-Value: {p_value}")

# Interpretation
if p_value < 0.05:
    print("The data is not normally distributed (reject H0).")
else:
    print("The data follows a normal distribution (fail to reject H0).")
```

Kesimpulan: Monetary tidak berdistribusi Normal

##Uji Hipotesis

```python
#uji non parametrik menggunakan spearman's rank correlation

from scipy.stats import spearmanr
# Menghitung Spearman's Rank Correlation
stat, p_value = spearmanr(df_marketing['Income'], df_marketing['Monetary'])

# Menampilkan hasil
print(f"Spearman's Rank Correlation Statistic: {stat}")
print(f"P-Value: {p_value}")

# Interpretasi
if p_value < 0.05:
    print("Ada korelasi signifikan antara Income dan Monetary.")
else:
    print("Tidak ada korelasi signifikan antara Income dan Monetary.")
```

Pvalue<=0,05
Kesimpulan:

Ada korelasi yang signifikan antara Income dan Monetary.
